## 📝 Overview
From the very first day I began my research journey at RUET, I dreamt of publishing in a **Q1 journal** before finishing my 3rd year of undergraduate studies. This work, which began as my **3-1 Urban Planning Studio** project, has officially become my first journal publication. After 14 months of dedication—8 months in review and 6 months of intensive work—this milestone was reached just before my 3-2 semester final exams.

Dhaka currently stands as one of the world's least liveable cities, plagued by overpopulation and ecological decline[cite: 5, 35, 37]. [cite_start]This study evaluates **Purbachal New Town**, the largest planned township in Bangladesh, which was designed to decentralize the primate city. By integrating **spatial analysis** and **qualitative governance frameworks**, I investigated the disconnect between ambitious master plans and ground implementation].

The project introduces the concept of **"Performative Decentralization"**—a phenomenon where decentralization is used as a policy rhetoric while implementation reinforces centralized power and elite interests. Using a multi-source approach, I triangulated data from official RAJUK documents, expert Key Informant Interviews (KIIs), and systematic validation with residents.

My findings reveal that while Purbachal follows international layout standards, its success is undermined by **elite capture** and a lack of local job integration. This research provides a replicable framework for urban planners in the Global South to evaluate the effectiveness of new town developments as instruments for genuine decentralization.

## ⚙️ Methodology
| Step | Description |
|------|-------------|
| 1. Data Collection | Triangulated official planning documents, 7 high-level expert KIIs, and 68 resident validation interviews. |
| 2. Preprocessing    | Audio recordings were transcribed, anonymized, and imported into **ATLAS.ti** for multi-layered analysis. |
| 3. Analysis         | Followed Grounded Theory to identify 11 themes, using axial coding to create 8 distinct code clusters. |
| 4. Visualization    | Developed Problem Trees, Sankey diagrams for causal links, and Heatmaps for thematic intensity. |

## 📊 Results & Insights
- 🔹 **Institutional Failures:** Identified widespread "elite capture" where prime plots are controlled by high-income beneficiaries, leading to the displacement of original landowners.
- 🔹 **Dormitory Town Effect:** Purbachal lacks a vibrant local economy, forcing 80% of residents to depend on Dhaka for jobs, which contradicts the goal of a self-sustained township.
- 🔹 **Planning Flaws:** The "design population" was revised from 1 million to 2 million (and potentially up to 4 million) without corresponding utility upgrades.

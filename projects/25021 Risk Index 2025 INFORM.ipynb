## üìù Overview
The **Global Risk Index 2025** visualizes how different countries face varying degrees of disaster risk based on the **INFORM Risk Index** ‚Äî a globally recognized framework that assesses risk by combining **hazard & exposure**, **vulnerability**, and **lack of coping capacity**.  

This Python-powered map highlights spatial patterns of global risk, identifying which regions are most susceptible to crises ‚Äî from natural disasters like floods and droughts to human-driven factors such as conflict or poor governance.

## ‚öôÔ∏è Methodology
| Step | Description |
|------|-------------|
| **1. Data Collection** | Acquired the *INFORM Risk Index 2025* dataset from **[HDX (Humanitarian Data Exchange)](https://data.humdata.org/)**, containing country-level indicators on risk, hazard, vulnerability, and coping capacity. |
| **2. Preprocessing** | Cleaned and selected key columns including `COUNTRY`, `ISO3`, `INFORM RISK`, `RISK CLASS`, and `Rank`. Standardized country names for accurate merging with Natural Earth shapefile. |
| **3. Spatial Join** | Merged the INFORM dataset with **Natural Earth Admin 0 shapefile** using ISO3 or country name for spatial mapping. |
| **4. Visualization** | Mapped countries using **GeoPandas** with the `YlOrRd` colormap to depict INFORM Risk Index values. Added legend, title, and source credits for professional presentation. |

## üìä Results & Insights
- üåã **Highest Risk (Very High INFORM Index):** South Sudan, Afghanistan, Somalia, Sudan, and Yemen.  
- üåæ **Medium Risk Regions:** India, Bangladesh, Indonesia, and parts of Latin America.  
- üïäÔ∏è **Low-Risk Nations:** Most of Western Europe, Japan, Canada, and Australia.  

The visualization reveals stark **spatial inequalities** in resilience and preparedness ‚Äî especially in regions experiencing conflict or fragile governance systems.
